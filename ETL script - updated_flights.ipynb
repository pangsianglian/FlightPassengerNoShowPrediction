### Vectorized Distance Calculation

```python
from numpy import radians, sin, cos, sqrt, arctan2

def haversine_vectorized(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * arctan2(sqrt(a), sqrt(1 - a))

    return R * c

# Apply to DataFrame
flights['Distance_km'] = (haversine_vectorized(
    flights['OriginLat'], flights['OriginLon'],
    flights['DestLat'], flights['DestLon']
)).astype(int)

# then drop the original Distance_Miles
flights = flights.drop('Distance_Miles', axis=1)
```

```python
# Merge New Origin coordinates
flights = flights.merge(
    airports.rename(columns={'IATA': 'Origin'}),
    on='Origin',
    how='left'
).rename(columns={'Latitude': 'Origin_Lat', 'Longitude': 'Origin_Lon'})

# drop the original unnecessary columns
flights = flights.drop(['OriginLat','OriginLon'], axis=1)
flights.head()
```

```python
# Define realistic airport list
dest_airportlist = ['ATL', 'DXB', 'DFW', 'HND', 'LHR', 'DEN', 'IST', 'ORD', 'DEL', 'PVG', 'NRT', 'KUL', 'SIN', 'ICN', 'SYD', 'BKK', 'HKG', 'TPE']
dest_airports = airports[airports['IATA'].isin(dest_airportlist)].copy()

# Find Closest Airport Matching Distance_km from New Origins
def find_matching_destination(Origin_Lat, Origin_Lon, Distance_km):
    def dist_diff(row):
        d = haversine_vectorized(Origin_Lat, Origin_Lon, row['Latitude'], row['Longitude'])
        return abs(d - Distance_km)
    

    closest = dest_airports.copy()
    closest['DistDiff'] = closest.apply(dist_diff, axis=1)
    best_match = closest.loc[closest['DistDiff'].idxmin()]
    return pd.Series([best_match['IATA'], best_match['Latitude'], best_match['Longitude']])

# Replace New Destinations, DestLon, DestLat dataset
flights[['Destination', 'DestLat', 'DestLon']] = flights.apply(
    lambda row: find_matching_destination(
        row['Origin_Lat'], row['Origin_Lon'], row['Distance_km']
    ),
    axis=1
)
```

```python
# Apply to DataFrame
flights['Distance_km'] = (haversine_vectorized(
    flights['Origin_Lat'], flights['Origin_Lon'],
    flights['DestLat'], flights['DestLon']
)).astype(int)

flights.head()
```

```python
import holidays

# 1. split out date and time
flights['DepartureDate'] = flights['Departure_Time'].dt.date

# 2. load Singapore’s holiday calendar
sg_holidays = holidays.Singapore(years=flights['DepartureDate'].apply(lambda d: d.year).unique().tolist())

# 3. flag rows whose date is in the holiday list
# .astype(int) - display IsHoliday into 0 (False - Non-holiday) and 1 (True - Holiday)
sg_ph = flights['DepartureDate'].isin(sg_holidays).astype(int) 
```

### Calculate IsHoliday (Global School holidays season)

```python
# — 2) Functions to compute U.S. school break ranges per year

def get_us_winter_break(year):
    # Christmas/Winter Break usually starts on the 3rd Saturday in December
    # and ends just after New Year’s Day :contentReference[oaicite:0]{index=0}
    dec = pd.date_range(f'{year}-12-01', f'{year}-12-31')
    third_sat = dec[dec.weekday == 5][2]
    # first Monday after Jan 2 of next year
    jan = pd.date_range(f'{year+1}-01-02', f'{year+1}-01-08')
    first_mon = jan[jan.weekday == 0][0]
    return third_sat.date(), first_mon.date()

def get_us_summer_break(year):
    # Summer Break ≈ 8–12 weeks from just after Memorial Day
    # through Labor Day :contentReference[oaicite:1]{index=1}
    may = pd.date_range(f'{year}-05-01', f'{year}-05-31')
    memorial = may[may.weekday == 0][-1]              # last Monday in May
    start = memorial + pd.Timedelta(days=1)           # the following day
    sept = pd.date_range(f'{year}-09-01', f'{year}-09-07')
    labor_day = sept[sept.weekday == 0][0]            # first Monday in Sept
    return start.date(), labor_day.date()

# — 3) China Spring Festival break (public holiday week)
# Chinese New Year public holiday: Jan 20–Feb 20 :contentReference[oaicite:2]{index=2}
def get_cn_spring_break(year):
    return (pd.to_datetime(f'{year}-01-20').date(),
            pd.to_datetime(f'{year}-02-20').date())

# — 4) Build a helper for “in range”
def in_range(d, r):  
    return r[0] <= d <= r[1]

# — 5) Precompute break ranges for all years in your data
years = flights['DepartureDate'].apply(lambda d: d.year).unique()
us_winter = {y: get_us_winter_break(y) for y in years}
us_summer = {y: get_us_summer_break(y) for y in years}
cn_spring = {y: get_cn_spring_break(y) for y in years}

# — 6) Flag each season and overall IsHoliday
flights['WinterHoliday'] = flights['DepartureDate'].apply(lambda d: in_range(d, us_winter[d.year]))
flights['SummerHoliday'] = flights['DepartureDate'].apply(lambda d: in_range(d, us_summer[d.year]))
flights['SpringHoliday'] = flights['DepartureDate'].apply(lambda d: in_range(d, cn_spring[d.year]))

# 1 = any of those breaks; 0 = regular session
sch_breaks = flights[['WinterHoliday','SummerHoliday','SpringHoliday']].any(axis=1)

```

```python
# combine SG public holidays and Global School Break
flights['IsHoliday'] = (sg_ph | sch_breaks).astype(int)

# then drop the original timestamp
flights = flights.drop(['WinterHoliday','SummerHoliday', 'SpringHoliday'], axis=1)
flights.shape
```
